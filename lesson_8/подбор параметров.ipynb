```python
import pandas as pd
import numpy as np

from sklearn.linear_model import Ridge

from sklearn.model_selection import KFold, GridSearchCV, RandomizedSearchCV
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('housing.csv')
```

```python
data
```

```python
data.info()
data = data.drop(["material", "sub_area", "radiation", "detention"], axis=1)
```

```python
data = data.fillna(0)
```

```python
train, test = train_test_split(data, test_size=0.2)

X_train=train.drop(["price_doc"], axis=1)
y_train=train["price_doc"]

X_test=test.drop(["price_doc"], axis=1)
y_test=test["price_doc"]
```

```python
grid = {'alpha': np.arange(1, 10_000, 100)}

cv = KFold(n_splits=5, shuffle=True, random_state=241)

clf = Ridge()
gs = GridSearchCV(clf, grid, scoring="r2", cv=cv, verbose=True)
gs.fit(X_train, y_train)
```

```python
bestA=gs.best_estimator_.alpha
```
